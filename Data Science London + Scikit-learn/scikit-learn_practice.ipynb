```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_x = pd.read_csv('train.csv')
train_y = pd.read_csv('trainLabels.csv')
test_x = pd.read_csv('test.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier

#scaler = StandardScaler()
#scaler.fit(train_x)
#train_x = scaler.transform(train_x)

model = LogisticRegression(C=100.0, max_iter=10000)
model.fit(train_x, train_y)
print('lr accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = DecisionTreeClassifier()
model.fit(train_x, train_y)
print('decison tree accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = svm.LinearSVC()
model.fit(train_x, train_y)
print('linear svm accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = svm.SVC()
model.fit(train_x, train_y)
print('rbf kernel svm accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = GaussianNB()
model.fit(train_x, train_y)
print('navie bayes accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(60), random_state=1)
model.fit(train_x, train_y)
print('nn accuracy: %f' % float(100 * model.score(train_x, train_y)))

model = RandomForestClassifier(n_estimators=100)
model.fit(train_x, train_y)
print('random forest accuracy: %f' % float(100 * model.score(train_x, train_y)))
```

```python
model = DecisionTreeClassifier()
model.fit(train_x, train_y)
print('decison tree accuracy: %f' % float(100 * model.score(train_x, train_y)))
test_y = pd.DataFrame(model.predict(test_x))
test_y.index = test_y.index + 1
```

```python
test_y.to_csv('dt_subbmission.csv')
```

```python
model = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(40,), max_iter=1000, random_state=1)
print(model)
split = StratifiedShuffleSplit(train_y, train_size=0.6)
for train_index, dev_index in split:
    features_train,features_dev=train_x.iloc[train_index],train_x.iloc[dev_index]
    labels_train,labels_dev=train_y.iloc[train_index],train_y.iloc[dev_index]
features_train.index=range(len(features_train))
features_dev.index=range(len(features_dev))
labels_train.index=range(len(labels_train))
labels_dev.index=range(len(labels_dev))
model.fit(features_train, labels_train)
print('nn train accuracy: %f' % float(100 * model.score(features_train, labels_train)))
print('nn dev accuracy: %f' % float(100 * model.score(features_dev, labels_dev)))
test_y = pd.DataFrame(model.predict(test_x))
test_y.index = test_y.index + 1
test_y.to_csv('nn_subbmission.csv')
```
